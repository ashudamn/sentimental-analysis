```python
stops=set(stopwords.words('english'))
stops.add("@united")
stops.add("@usairway")
stops.add("@americanair")
stops.add("@southwest")
stops.add("@delta")
stops.add("@virginamerica")  
def cleanData(text):
    meaningful_word=[]
    text=text.split()
    for w in text:
        if w.lower() not in stops:
            meaningful_word.append(w.lower())   
    return (" ".join(meaningful_word))    
```

```python
tweets['clean_tweets']=[cleanData(txt) for txt in tweets['text']]
```

```python
tweets['clean_tweets']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v=CountVectorizer(analyzer='word',ngram_range=(1,2))
```

```python
from sklearn import cross_validation
```

```python
X_train,X_test,Y_train,Y_test=cross_validation.train_test_split(tweets['clean_tweets'],tweets['sentiment'],test_size=0.3)
```

```python
train_features=v.fit_transform(X_train)
test_features=v.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf=MultinomialNB()
clf.fit(train_features,Y_train)
clf.score(test_features,Y_test)
```
